# Pregunta 2. 

Desarrolla una función capaz de recortar y pegar un fragmento rectangular de una imagen en otra ubicación (de la misma imagen). Para esto, la función debería recibir como parámetros de entrada: la imagen (I), las coordenadas de la esquina superior izquierda del rectángulo (x_origen, y_origen), su ancho (w) y alto (h). Finalmente, la función también recibe las coordenadas de destino (x_destino, y_destino), que indican donde se pegará el fragmento (esquina superior izquierda).

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def recortar_y_pegar(imagen, x_origen, y_origen, w, h, x_destino, y_destino):
    imagen_modificada = imagen.copy()
    fragmento = imagen[y_origen:y_origen+h, x_origen:x_origen+w].copy()
    imagen_modificada[y_destino:y_destino+h, x_destino:x_destino+w] = fragmento

    return imagen_modificada
    
imagen = cv2.imread('./images/peppers.png')
imagen_rgb = cv2.cvtColor(imagen, cv2.COLOR_BGR2RGB)

x_origen, y_origen, w, h = 100, 50, 50, 50   
x_destino, y_destino = 200, 150              

imagen_modificada = recortar_y_pegar(imagen_rgb, x_origen, y_origen, w, h, x_destino, y_destino)

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(imagen_rgb)
plt.title("Imagen Original")
plt.axis("off")

plt.subplot(1, 2, 2)
plt.imshow(imagen_modificada)
plt.title("Imagen Modificada")
plt.axis("off")

plt.show()
```
